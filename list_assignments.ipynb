```python
## Extract all 'name' values from list of dictionaries"
dic=[{'name': 'Alice', 'age': 25},
    {'name': 'Bob', 'age': 30}
]
names=[]
for i in dic:
    names.append(i['name'])
print(names)
```

```python
## Find the person with maximum age in list of dictionaries
people=[{'name': 'Alice', 'age': 25},
    {'name': 'Bob', 'age': 30}]
max_person = people[0]

for person in people:
    if person['age'] > max_person['age']:   # dictionary indexing operator
        max_person = person

print(max_person)
```
