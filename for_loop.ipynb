```python
##Count how many prime numbers exist between 1 and n.

def count_prime(n):
    count=0
    for num in range(2,n+1):
        is_prime=True

        for i in range(2,num):
            if num%i==0:
                is_prime=False
                break
        if is_prime:
            count+=1
    print(count)
count_prime(100)
```

```python
##Calculate sum of harmonic series: 1 + 1/2 + 1/3 + ... + 1/n
def harmonic_series(n):
    sum=0
    for i in range(1,n+1):
        sum+=1/i
    print(sum) 
harmonic_series(10)  
```

```python
#Reverse a string using for loop.
def reverse_string(s):
    reversed_str=""
    for ch in range(len(s)-1,-1,-1):
        reversed_str+=s[ch]
    print(reversed_str)
reverse_string("darshan")
```

```python
##Count frequency of a specific digit in a number.
def frequency_of_digit(num,digit):
    count=0
    for i in str(num):
        if i==str(digit):
            count+=1
    print(count)
frequency_of_digit(123334,3)
```

```python
## Calculate: 1 - 2 + 3 - 4 + 5 - 6 + ... ± n
def k(n):
    sum=0
    for i in range(1,n+1):
        if i%2!=0:
            sum+=i
        else:
            sum-=i
    print(sum)
k(100)
```

```python
##Check if a string is a palindrome using for loop.
def palindrome(s):
    reverse_str=""
    for i in range(len(s)-1,-1,-1):
        reverse_str+=s[i]
    if s==reverse_str:
        print("string is palindrome")
    else:
        print("string is not a palindrome")

palindrome("madam")
```

```python
##Convert a binary number (as string) to decimal using for loop.
def binary_to_decimal(b):
    decimal=0
    power=0
    for i in range(len(b)-1,-1,-1):
        decimal+=int(b[i])*(2**power)
        power+=1
    print(decimal)
binary_to_decimal("1011")
```

```python
##Calculate compound interest for n years: A = P(1 + r/100)^t
P = float(input("Enter principal amount: "))
r = float(input("Enter rate of interest: "))
t = int(input("Enter number of years: "))

A = P * (1 + r/100) ** t
CI = A - P

print(f"Amount ={A}")
print(f"Compound Interest ={CI}")
```

```python
##Print all perfect numbers between 1 and n.
def list_of_prime(n):
    list=[]
    for i in range(2,n):
        for j in range(2,i):
            if i%j==0:
                break
        else:
            list.append(i)
    print(list)
list_of_prime(100)
```

```python
##Calculate: 1! + 2! + 3! + ... + n!
def nested_fact(n):
    
    sum=0
    for i in range(1,n+1):
        fact=1
        for j in range(1,i+1):
            fact*=j
        sum+=fact
    
    print(sum)
nested_fact(5)
```
