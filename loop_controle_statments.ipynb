```python
##Write a program to check if a number is an Armstrong number (also called Narcissistic number).
n=int(input("enter the number"))
original=n
n_str=str(n)
power=len(n_str)
total=0

for i in n_str:
    total+=int(i)**power

if total==original:
    print("armstrong number")
else:
    print("not a armstrong number")
```

```python
##Print a diamond pattern for n rows (n should be odd):
n=int(input("enter the odd number of rows"))
if n%2==0:
    print("please enter odd number of rows")
else:
    mid=n//2+1
    for i in range(1,mid+1):
        for j in range(mid-i):
            print(" ",end=" ")
        for k in range(2*i-1):
            print("*",end=" ")
        print()

    for i in range(mid-1,0,-1):
        for j in range(mid-i):
            print(" ",end=" ")
        for k in range(2*i-1):
            print("*",end=" ")
        print()
```

```python
##Write a program to check if a number is a perfect number.

num=abs(int(input("enter the number")))
sum=0
if num>0:
    for i in range(1,num):
        if num%i==0:
            sum+=i
    if sum==num:
        print(f"{num} is perfect number")
    else:
        print(f"{num} is not a perfect number")
```

```python
##Write a program to find the Least Common Multiple (LCM) of two numbers.
def gcd(num1,num2):
    while num2!=0:
        reminder=num1%num2
        num1=num2
        num2=reminder
    return num1
## LCM

def lcm_calculator(a,b):
    lcm=(a*b)//gcd(a,b)
    return lcm
lcm_calculator(34,24)
```

```python
##Write a program to check if a number is a Strong number.
def strong_number(num):
    originale=num
    sum=0
    
    while originale>0:
        digit=originale%10
        factorial_of_digit=1

        for j in range(1,digit+1):
            factorial_of_digit*=j

        sum+=factorial_of_digit
        originale//=10

    if sum==num:
        print(f"{num} is strong number")
    else:
        print(f"{num} is not a strong number")
strong_number(145)
    





```

```python
##Create a simple calculator program that:Shows a menu of operations: Add, Subtract, Multiply, Divide, Exit 2 Takes two numbers as input 3 Performs the selected operation 4 Shows the result 5 Repeats until the user chooses to exit 6 Handles division by zero error

def simple_calculator():
    while True:
        print("-----welcome to simple calculator-----")
        print("1:ADD")
        print("2:SUBSTRACT")
        print("3:MULTIPLY")
        print("4:DIVITION")
        print("5:EXIT")

        choice =input("please enter your choice (1-5):")

        if choice=='5':
            print(f"exiting calculator")
            break

        if choice in ["1",'2','3','4','5']:
            num1=int(input("enter the value"))
            num2=int(input("enter the value"))

            if choice == '1':
                print("result:",num1+num2)
            elif choice == '2':
                print("result:",num1-num2)
            elif choice == '3':
                print("result:",num1*num2)
            elif choice == '4':
                try:
                    print("result:",num1/num2)
                except ZeroDivisionError:
                    print("number is not divisible by zero, please enter positive number")
        else:
            print("invalid choice")
simple_calculator()
```

```python
##The Collatz conjecture is one of the most famous unsolved problems in mathematics.

def collatz_conjecture(n):
    count=0
    print(n,end=" ")

    while n!=1:
        if n%2==0:
            n=n//2
            
        else:
            n=3*n+1
        print(n,end=" ")
        count+=1
    print("\ntotal steps",count)
collatz_conjecture(6)
```

```python
##Create an ATM machine simulator with the following features:

pin=1234
balance=10000
attempts=0
while attempts<3:
    enter_pin=int(input("enter the pin"))
    if pin==enter_pin:
        print("login successful")

        while True:
            print("---user menu----")
            print("1:check balance")
            print("2:withdraw money")
            print("3:deposite money")
            print("4:change pin")
            print("5:exit")

            choice=input("enter your choice(1-5)")

            if choice=='5':
                print("thank you for using ATM , visit again...")
                break

            if choice in ['1','2','3','4']:
                if choice == '1':
                    print("current balance:",balance)
                elif choice == '2':
                    withdraw_amount=float(input("enter amount to withdraw"))
                    if withdraw_amount>balance:
                        print("low balance")
                    else:
                        balance-=withdraw_amount
                        print("please collect amount ")
                        print("remaining balance:",balance)
                elif choice == '3':
                    deposite_amount=float(input("enter the deposite amount"))
                    balance+=deposite_amount
                    print("amount deposited")
                    print(f"total balnce:{balance}")

                elif choice =='4':
                    new_pin=int(input("enter new pin"))
                    pin=new_pin
                    print("pin changed successfully.")
                else:
                    print("invalid choice")
            break

    else:
        attempts+=1
        print(f"wrong pin , attempts left:{3-attempts}")
if attempts==3:
    print("too many Attempts, account blocked")
```
