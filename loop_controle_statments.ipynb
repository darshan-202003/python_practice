```python
##Write a program to calculate the factorial of a number using a for loop.
num=int(input("enter the number:"))
fact=1
if num<0:
    print("enter positive number")
else:
    for i in range(1,num+1):
        fact*=i
    print(f"Factorial of {num} is {fact}")
```

```python
##Write a program to print the first n terms of the Fibonacci series.
n=int(input("enter the value of n"))
a=0
b=1
list=[]
if n<=0:
    print("please enter a positive number")
elif n==1:
    list.append(a)
    list.append(b)
else:
    list.append(a)
    list.append(b)
    for i in range(2,n):
        c=a+b
        list.append(c)
        a=b
        b=c
print(list)
```

```python
##Write a program to check if a given number is prime or not.
n=int(input("enter the value of n"))
if n<=1:
    print(f"{n} is not a prime number")
else:
    for i in range(2,n-1):
        if n%i==0:
            print(f"{n}  is not a prime number")
    else:
        print(f"{n}  is a prime number")
         
```

```python
##Write a program to check if a given number is a palindrome or not.
n=121
temp=n
reverse=0
while temp>0:
    digit=temp%10
    reverse=reverse*10+digit
    temp=temp//10

if n==reverse:
    print('IT IS PALINDROME')
else:
    print("IT IS NOT PALINDROME")
```

```python
##Write a program to check if a number is an Armstrong number (also called Narcissistic number).
n=int(input("enter the number"))
original=n
n_str=str(n)
power=len(n_str)
total=0

for i in n_str:
    total+=int(i)**power

if total==original:
    print("armstrong number")
else:
    print("not a armstrong number")
```

```python
##Print a diamond pattern for n rows (n should be odd):
n=int(input("enter the odd number of rows"))
if n%2==0:
    print("please enter odd number of rows")
else:
    mid=n//2+1
    for i in range(1,mid+1):
        for j in range(mid-i):
            print(" ",end=" ")
        for k in range(2*i-1):
            print("*",end=" ")
        print()

    for i in range(mid-1,0,-1):
        for j in range(mid-i):
            print(" ",end=" ")
        for k in range(2*i-1):
            print("*",end=" ")
        print()
```
