```python
# Question 4: Find the intersection of {1, 2, 3, 4, 5} and {4, 5, 6, 7, 8}
set1={1, 2, 3, 4, 5}
set2={4, 5, 6, 7, 8}
intersection=set1.intersection(set2)
print(intersection)
```

```python
# Question 6: Check if 5 is in set {1, 2, 3, 4, 6, 7, 8}
set= {1, 2, 3, 4, 6, 7, 8}
find=5
for i in set:
    if i==find:
        print(True)
else:
    print(False)
```

```python
# Question 7: Find the length of set {'a', 'b', 'c', 'd', 'e'}
set={'a', 'b', 'c', 'd', 'e'}
count=0
for i in set:
    count+=1
print(count)
```

```python
# Question 8: Create a set of vowels from string "hello world"
string="hello world"
vowels="aeiou"
l=""
for i in string:
    if i in vowels:
        l+=i
print(set(l))
```
