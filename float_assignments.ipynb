```python
## Calculate the average of 3.14, 2.718, 1.618, 0.577
num=[3.14, 2.718, 1.618, 0.577]
avg=sum(num)/len(num)
print(avg)
```

```python
## Calculate compound interest on $1000 at 5.5% for 3 years
principal=1000
rate=5.5/100
time=3
compound=round(principal*(1+rate)**time,3)
print(compound)
```

```python
## Find the hypotenuse of a right triangle with sides 3.5 and 4.2
import math
a=3.5
b=4.2
hypotenuse=math.sqrt(a**2+b**2)
print(hypotenuse)
```
