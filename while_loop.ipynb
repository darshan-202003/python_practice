```python
##Print Fibonacci numbers until they exceed n.
def fib(n):
    a=0
    b=1

    while a<=n:
        print(a,end=' ')
        c=a+b
        a=b
        b=c
fib(10)


```

```python
##A neon number is one where sum of digits of its square equals the number. Example: 9 (9² = 81, 8+1 = 9)
n=7
square=n*n
sum=0
while square >0:
    digit=square%10
    sum+=digit
    square=square//10

if sum==n:
    print("neon number")
else:
    print("not a neon number")
```

```python
##A spy number is one where sum of digits equals product of digits. Example: 123 (1+2+3 = 6, 1×2×3 = 6)
n=123
product=1
sum=0
while n>0:
    digit=n%10
    sum+=digit
    product*=digit
    n=n//10
print(sum)
print(product)
if sum==product:
    print("it's a spy number")
else:
    print("it's not a spy number")
```

```python
##A Harshad number is divisible by the sum of its digits. Example: 18 (1+8=9, 18÷9=2)
n=18
num=n
sum=0
while n>0:
    digit=n%10
    sum+=digit
    n=n//10
if num%sum==0:
    print("it's a harshad number")
else:
    print("it's not a harshad number")
```

```python
##An abundant number is one where sum of proper divisors exceeds the number. Example: 12 (1+2+3+4+6 = 16 > 12)
num=12
n=num
i=1
sum=0
while i<num:
    if n%i==0:
        sum+=i
    i+=1
if sum>num:
    print("it's abundant number")
else:
    print("it's not abundant number")
```

```python
##A deficient number is one where sum of proper divisors is less than the number. Example: 8 (1+2+4 = 7 < 8)
num=15
n=num
i=1
sum=0
while i<num:
    if n%i==0:
        sum+=i
    i+=1
if sum<num:
    print("it's deficient number")
else:
    print("it's not deficient number")
```

```python
##Check if two numbers are amicable (sum of proper divisors of each equals the other). Example: 220 and 284
num1=220
num2=284
i=1
sum1=0
while i<num1:
    if num1%i==0:
        sum1+=i
    i+=1

j=1
sum2=0
while j<num2:
    if num2%j==0:
        sum2+=j
    j+=1

if num1==sum2 and num2==sum1:
    print("Amicable number")

else:
    print("not amicable number")
```

```python
##A magic number is one where repeatedly summing digits eventually gives 1. Example: 19 → 1+9=10 → 1+0=1
num=19
while num>9:
    sum=0
    while num>0:
        digit=num%10
        sum+=digit
        num=num//10
    num=sum

if num==1:
    print("magic number")

else:
    print("not magic number")
```

```python
##Validate a 10-digit ISBN number using the check digit algorithm.
isbn = input("Enter 10 digit ISBN: ")
i = 0
sum = 0

while i < 10:
    digit = int(isbn[i])
    sum += (i + 1) * digit
    i += 1

if sum % 11 == 0:
    print("Valid ISBN")
else:
    print("Invalid ISBN")
```
