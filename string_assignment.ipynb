```python
## Check if two strings are anagrams: 'listen' and 'silent'

word1="listen"
word2="silent"

word1=word1.replace(" ","").lower()
word2=word2.replace(" ","").lower()

if len(word1)!=len(word2):
    print("not anagram")
else:
    if sorted(word1)== sorted(word2):
        print("anagram")

    else:
        print("not anagram")
        
```

```python
##  Convert 'hello world' to 'hElLo WoRlD' (alternating case)
text = "hello world"
result = ""

for i in range(len(text)):
    if i % 2 == 0:
        result += text[i].lower()
    else:
        result += text[i].upper()

print(result)
```
